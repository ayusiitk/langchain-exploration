````python
from langchain import OpenAI

# The vectorstore we'll be using
from langchain.vectorstores import FAISS

# The LangChain component we'll use to get the documents
from langchain.chains import RetrievalQA

# The easy document loader for text
from langchain.document_loaders import TextLoader

# The embedding engine that will convert our text to vectors
from langchain.embeddings.openai import OpenAIEmbeddings

llm = OpenAI(temperature=1, openai_api_key=openai_api_key)
````

````python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
````

````python
import os
root_dir = data_dir
docs = []

# Go through each folder
for dirpath, dirnames, filenames in os.walk(root_dir):
    
    # Go through each file
    for file in filenames:
        try: 
            # Load up the file as a doc and split
            if(".txt" in file):
                loader = TextLoader(os.path.join(dirpath, file), encoding='utf-8')
            elif (".pdf" in file):
                loader = PyPDFLoader(os.path.join(dirpath, file))
            else:
                loader = TextLoader(os.path.join(dirpath, file), encoding='utf-8')
            doc = loader.load()
            text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=400)
            docs.extend(text_splitter.split_documents(doc))
        except Exception as e: 
            print(e)
````

````python
# Get your embeddings engine ready
embeddings = OpenAIEmbeddings(openai_api_key=openai_api_key)

# Embed your documents and combine with the raw text in a pseudo db. Note: This will make an API call to OpenAI
docsearch = FAISS.from_documents(docs, embeddings)
````

````python
qa = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=docsearch.as_retriever())
````

````python
query = "tell me major risks that mankind faces?"
qa.run(query)
````

````python
import PyPDF2
import docx
import pandas as pd
import pytesseract
from PIL import Image

def read_code_file(file_path):
    with open(file_path, 'r') as file:
        code_content = file.read()
    return code_content

def read_pdf_file(file_path):
    pdf_content = ""
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            pdf_content += page.extract_text()
    return pdf_content


def read_word_file(file_path):
    doc = docx.Document(file_path)
    word_content = ' '.join([paragraph.text for paragraph in doc.paragraphs])
    return word_content

def read_csv_file(file_path):
    df = pd.read_csv(file_path)
    csv_content = df.to_string(index=False)
    return csv_content

def read_scanned_pdf(file_path):
    pdf_content = ""
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfFileReader(file)
        num_pages = pdf_reader.numPages
        for page_num in range(num_pages):
            page = pdf_reader.getPage(page_num)
            image = page.to_image(resolution=300)
            text = pytesseract.image_to_string(image)
            pdf_content += text
    return pdf_content

# Example usage
file_path = r'C:\Users\ayush\Documents\workspace\langchain-tutorials\userdata\data\legal\SecureDocs_Archive8-4-2021\SecureDocs_Archive8-4-2021\Customers\HUL.pdf'

if file_path.endswith('.py'):
    content = read_code_file(file_path)
elif file_path.endswith('.pdf'):
    content = read_scanned_pdf(file_path)
elif file_path.endswith('.docx'):
    content = read_word_file(file_path)
elif file_path.endswith('.csv'):
    content = read_csv_file(file_path)
else:
    print("Unsupported file type.")

print(content)
````

````python
import cv2
import numpy as np

def preprocess_image(image):
    # Convert the PIL Image to a numpy array
    image = np.array(image)
    
    # Convert the image to grayscale
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply adaptive thresholding to binarize the image
    image = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

    # Optional: Apply dilation and erosion to remove some noise
    kernel = np.ones((1, 1), np.uint8)
    image = cv2.dilate(image, kernel, iterations=1)
    image = cv2.erode(image, kernel, iterations=1)

    return image
````

````python
import PyPDF2
import docx
import pandas as pd
import pytesseract
from PIL import Image

def read_code_file(file_path):
    with open(file_path, 'r') as file:
        code_content = file.read()
    return code_content

def read_pdf_file(file_path):
    pdf_content = ""
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            pdf_content += page.extract_text()
    return pdf_content

def read_word_file(file_path):
    doc = docx.Document(file_path)
    word_content = ' '.join([paragraph.text for paragraph in doc.paragraphs])
    return word_content

def read_csv_file(file_path):
    df = pd.read_csv(file_path)
    csv_content = df.to_string(index=False)
    return csv_content

def read_scanned_pdf(file_path):
    pdf_content = ""
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        num_pages = len(pdf_reader.pages)
        for page_num in range(num_pages):
            page = pdf_reader.pages[page_num]
            images = page.extract_images()
            for image in images:
                img = Image.open(image)
                text = pytesseract.image_to_string(img)
                pdf_content += text
    return pdf_content

# Example usage
file_path = 'path_to_your_file'

if file_path.endswith('.py'):
    content = read_code_file(file_path)
elif file_path.endswith('.pdf'):
    try:
        content = read_pdf_file(file_path)
    except PyPDF2.utils.PdfReadError:
        content = read_scanned_pdf(file_path)
elif file_path.endswith('.docx'):
    content = read_word_file(file_path)
elif file_path.endswith('.csv'):
    content = read_csv_file(file_path)
else:
    print("Unsupported file type.")

print(content)
````
